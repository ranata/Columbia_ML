```python
from __future__ import division
import numpy as np
import sys
```

```python
X_train = np.genfromtxt('X_train_C.csv', delimiter=',')
y_train = np.genfromtxt('Y_train_C.csv', delimiter=',')
X_test = np.genfromtxt('X_test_C.csv', delimiter=',')
#y_test = np.genfromtxt('Y_test_C.csv', delimiter=',')}
```

```python
print(X_train)
print(y_train)
```

```python
Y, n_y = np.unique(y_train, return_counts=True)
```

```python
print(Y, n_y)
```

```python
pi_cap = []
mu_cap = []
sigma_cap = []
f_x_cap = []
for y, count_y in zip(Y, n_y):
    
    pi_cap_y = count_y/np.sum(n_y)
    pi_cap.append(pi_cap_y)
#    print("pi_cap", pi_cap)
    
    mu_cap_y = np.dot((y_train == y), X_train)/count_y
    mu_cap.append(mu_cap_y)
    
    
    x_train_mu_cap_y_diff = X_train - mu_cap_y
    x_train_mu_cap_y_diff = (x_train_mu_cap_y_diff.T * (y_train == y)).T
    
    sigma_cap_y = np.dot(x_train_mu_cap_y_diff.T, x_train_mu_cap_y_diff)/count_y
    print("*****************************************")
    print(np.linalg.det(sigma_cap_y))
    
    sigma_cap.append(sigma_cap_y)
    
```

```python
# predicting for X_test
y_test = []
    
for x in X_test:
    print("x test - ", x)
    
    for y in Y:
        idx = np.int(y)
        x_mu_cap_y_diff = x - mu_cap[idx]

        coefficient = pi_cap[idx]/np.sqrt(np.linalg.det(sigma_cap[idx]))
        
        power_value = -0.5*np.dot(x_mu_cap_y_diff.T, np.dot(np.linalg.inv(sigma_cap[idx]), x_mu_cap_y_diff))
        
        print("y - ", y)
        
        print("****************")

        print("pi_cap[idx] - ", pi_cap[idx])
        print("coefficient - ", coefficient)
        
        print("power_value - ", power_value)
        print("exponent - ", np.exp(power_value))

        print("pdf value - ", coefficient*np.exp(power_value))

        f_x_cap_y = coefficient*np.exp(power_value)
        y_test.append(f_x_cap_y)

    f_x_cap.append(y_test)
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
qda = QuadraticDiscriminantAnalysis(store_covariance=True)
model = qda.fit(X_train, y_train)
```

```python
#print(model.priors_)
print(model.covariance_)
```

```python
pred=model.predict(X_test)
```
